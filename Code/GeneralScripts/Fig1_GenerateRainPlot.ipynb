```python
%matplotlib inline
import pandas as pd
import ptitprince as pt
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
DataDir = ('/home/ttha0011/kg98/Thapa/cTBS_Study/8_Figures/')

CMF_df = pd.read_csv(os.path.join(DataDir, 'ALFF/CMF_design_ALFFall.csv'))
RMF_df = pd.read_csv(os.path.join(DataDir, 'ALFF/RMF_design_ALFFall.csv'))
SF_df = pd.read_csv(os.path.join(DataDir, 'ALFF/SF_design_ALFFall.csv'))
GABA_df = pd.read_csv(os.path.join(DataDir, 'MRS/GABAall.csv'))

CMF_df.head()
```

```python
plt.rcParams["figure.figsize"] = (15,10)

plt.subplot(1,6,1)
ax1=sns.boxplot(x = "Group", 
                y = "CMF", 
                hue = "Time", 
                data = CMF_df,                 
                palette = "PuBu",
                dodge = True,  
                orient = "v",
                color = "black",
                width = 0.3)
                
ax1 = sns.pointplot(x = "Group",
                   y = "CMF", 
                   hue = "Time",
                   data = CMF_df, 
                   palette = "PuBu",
                   dodge=True,
                   markers=["o", "x"],
                   linestyles=["-", "--"])

sns.stripplot(x = "Group",
              y = "CMF", 
              hue = "Time", 
              data = CMF_df,
              size = 6,
              palette = "PuBu",           
              linewidth=0)
          

ax1.spines['right'].set_visible(False)
ax1.spines['top'].set_visible(False)

# Get the handles and labels
handles, labels = ax1.get_legend_handles_labels()

# When creating the legend, only use the first two elements to effectively remove the last two.
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.subplot(1,6,3)
ax2=sns.boxplot(x = "Group", 
                y = "RMF", 
                hue = "Time", 
                data = RMF_df,                 
                palette = "PuBu",
                dodge = True,  
                orient = "v",
                color = "black",
                width = 0.3)
                
ax2 = sns.pointplot(x = "Group",
                   y = "RMF", 
                   hue = "Time",
                   data = RMF_df, 
                   palette = "PuBu",
                   dodge=True,
                   markers=["o", "x"],
                   linestyles=["-", "--"])

sns.stripplot(x = "Group",
              y = "RMF", 
              hue = "Time", 
              data = RMF_df,
              size = 6,
              palette = "PuBu",           
              linewidth=0)
          

ax2.spines['right'].set_visible(False)
ax2.spines['top'].set_visible(False)

# Get the handles and labels
handles, labels = ax2.get_legend_handles_labels()

# When creating the legend, only use the first two elements to effectively remove the last two.
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


plt.subplot(1,6,5)
ax3=sns.boxplot(x = "Group", 
                y = "SF", 
                hue = "Time", 
                data = SF_df,                 
                palette = "PuBu",
                dodge = True,  
                orient = "v",
                color = "black",
                width = 0.3)
                
ax3 = sns.pointplot(x = "Group",
                   y = "SF", 
                   hue = "Time",
                   data = SF_df, 
                   palette = "PuBu",
                   dodge=True,
                   markers=["o", "x"],
                   linestyles=["-", "--"])

sns.stripplot(x = "Group",
              y = "SF", 
              hue = "Time", 
              data = SF_df,
              size = 6,
              palette = "PuBu",           
              linewidth=0)
          

ax3.spines['right'].set_visible(False)
ax3.spines['top'].set_visible(False)

# Get the handles and labels
handles, labels = ax3.get_legend_handles_labels()

# When creating the legend, only use the first two elements to effectively remove the last two.
l = plt.legend(handles[0:2], labels[0:2], bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)


#plt.tight_layout()
plt.savefig(os.path.join(DataDir, 'ALFF.png'))
```
